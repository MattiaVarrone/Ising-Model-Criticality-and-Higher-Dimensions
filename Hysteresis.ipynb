```python
import numpy as np
from scipy.optimize import curve_fit
import matplotlib.pyplot as plt
from Ising import Lattice
from spec_heat import C
```

```python
l = Lattice(100, 2)
T = 1
ns = [5, 50]
fig, ax = plt.subplots()
for n in ns:
    l.random()
    Hs, magns = l.hysteresys(n, T, H_max=2, H_steps=30, cycles=10)
    ax.plot(Hs, magns, linestyle='dotted')

ax.set_xlabel("$H$")
ax.set_ylabel("$M$")
ax.set_title(f"Hysteresis with 10 cyles, {T = }, (N = 100)")
ax.legend([f'no.sweeps = {n}' for n in ns], loc='upper left')
fig.set_size_inches(9, 7)
fig.savefig(r"pics/hysteresis" + str(T) + ".png", format="png", dpi=1200, bbox_inches='tight')
```
